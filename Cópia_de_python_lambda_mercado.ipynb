## 📦 SEÇÃO 1 — O Caixa Eletrônico
### Conceito: Lambda básica — o que é e como funciona

**Situação:** O sistema de caixa precisa calcular o preço final de um produto  
aplicando um percentual de desconto. O desenvolvedor anterior usava funções  
enormes para operações simples. Você vai simplificar com lambda.

---

### 📖 Teoria — Leia antes de executar

Uma **lambda** é uma função sem nome, escrita em uma linha:

```
lambda parâmetros : expressão
    ↑                  ↑
  entradas        o que retorna
```

É exatamente igual a uma `def`, mas compacta:

```python
# Com def (jeito tradicional)
def desconto(preco, pct):
    return preco - (preco * pct / 100)

# Com lambda (equivalente)
desconto = lambda preco, pct: preco - (preco * pct / 100)
```

````python
# ── DEMONSTRAÇÃO 1A ── Lambda básica vs def ──────────────────────────

# Jeito antigo (def)
def desconto_def(preco, pct):
    return preco - (preco * pct / 100)

# Jeito novo (lambda)
desconto_lambda = lambda preco, pct: preco - (preco * pct / 100)

preco_leite = 8.50

print("🥛 Produto: Leite integral — R$ 8,50")
print(f"   def    → R$ {desconto_def(preco_leite, 15):.2f}  (15% off)")
print(f"   lambda → R$ {desconto_lambda(preco_leite, 15):.2f}  (15% off)")
print()
print("✅ Mesmo resultado! Lambda é só uma forma mais curta de escrever.")
````

````python
# ── DEMONSTRAÇÃO 1B ── Lambda com 1, 2 e 3 parâmetros ────────────────

# 1 parâmetro — verificar se produto está em promoção
em_promocao = lambda preco: preco < 5.00

# 2 parâmetros — calcular preço com desconto
preco_final = lambda preco, desconto: preco * (1 - desconto / 100)

# 3 parâmetros — calcular total com desconto e frete
total_compra = lambda preco, desconto, frete: preco * (1 - desconto / 100) + frete

print("🏷️  Produto: Biscoito — R$ 3,99")
print(f"   Em promoção?      {em_promocao(3.99)}")
print(f"   Com 10% desconto: R$ {preco_final(3.99, 10):.2f}")
print(f"   + R$ 2,00 frete:  R$ {total_compra(3.99, 10, 2.00):.2f}")
````

## 📦 SEÇÃO 2 — Organizando o Estoque
### Conceito: Lambda como `key=` no `sorted()`

**Situação:** O estoquista precisa organizar os produtos por data de validade  
para colocar os que vencem mais cedo na frente da prateleira.  
Os produtos estão numa lista de dicionários — como ordenar por um campo específico?

---

### 📖 Teoria

`sorted()` aceita um parâmetro `key=` que define **por qual critério** ordenar.  
Lambda é perfeita aqui porque você quer uma regra simples e pontual:

```python
sorted(lista, key=lambda item: item['campo'])
#                  ↑ cada elemento    ↑ critério de comparação
```

````python
# ── DEMONSTRAÇÃO 2A ── Ordenar por validade ───────────────────────────

estoque = [
    {'produto': 'Iogurte',      'validade': date(2025, 6, 3),  'preco': 4.50},
    {'produto': 'Queijo Minas', 'validade': date(2025, 5, 15), 'preco': 12.90},
    {'produto': 'Manteiga',     'validade': date(2025, 8, 20), 'preco': 9.80},
    {'produto': 'Requeijão',    'validade': date(2025, 5, 28), 'preco': 7.30},
    {'produto': 'Creme de leite','validade': date(2025, 7, 10), 'preco': 5.20},
]

# Ordenar por validade (mais próximo primeiro)
por_validade = sorted(estoque, key=lambda p: p['validade'])

print("🧀 ESTOQUE ORDENADO POR VALIDADE (frente → fundo da prateleira)")
print("-" * 55)
for i, p in enumerate(por_validade, 1):
    print(f"  {i}. {p['produto']:18} vence em {p['validade'].strftime('%d/%m/%Y')}")

print()
print("⚠️  Atenção:", por_validade[0]['produto'], "vence primeiro!")
````

````python
# ── DEMONSTRAÇÃO 2B ── Ordenar por preço e por nome ───────────────────

# Do mais barato ao mais caro
por_preco = sorted(estoque, key=lambda p: p['preco'])

# Por ordem alfabética
por_nome = sorted(estoque, key=lambda p: p['produto'])

# Do mais caro ao mais barato (reverse=True)
mais_caro_primeiro = sorted(estoque, key=lambda p: p['preco'], reverse=True)

print("💰 DO MAIS BARATO AO MAIS CARO")
for p in por_preco:
    print(f"   R$ {p['preco']:.2f}  →  {p['produto']}")

print()
print("🔤 ORDEM ALFABÉTICA")
for p in por_nome:
    print(f"   {p['produto']}")
````

## 📦 SEÇÃO 3 — Fiscalização de Validade
### Conceito: Lambda com `filter()`

**Situação:** A vigilância sanitária vai fazer uma visita amanhã.  
O gerente precisa separar **urgentemente** os produtos que vencem nos próximos 7 dias  
para retirá-los das prateleiras antes da fiscalização.

---

### 📖 Teoria

`filter(função, lista)` percorre a lista e **mantém só os elementos** onde a função retorna `True`.

```python
filter(lambda x: condição_verdadeira, lista)
#              ↑ se True, o item fica
#                se False, o item é descartado
```

> **Analogia:** filter é como uma peneira. Lambda define o tamanho dos furos.

````python
# ── DEMONSTRAÇÃO 3A ── Produtos vencendo em breve ─────────────────────
from datetime import timedelta

hoje = date(2025, 5, 12)

produtos_loja = [
    {'produto': 'Leite',        'validade': date(2025, 5, 14), 'setor': 'Laticínios'},
    {'produto': 'Arroz',        'validade': date(2025, 12, 1), 'setor': 'Secos'},
    {'produto': 'Iogurte',      'validade': date(2025, 5, 16), 'setor': 'Laticínios'},
    {'produto': 'Feijão',       'validade': date(2025, 11, 5), 'setor': 'Secos'},
    {'produto': 'Presunto',     'validade': date(2025, 5, 13), 'setor': 'Frios'},
    {'produto': 'Queijo',       'validade': date(2025, 5, 19), 'setor': 'Frios'},
    {'produto': 'Macarrão',     'validade': date(2025, 9, 30), 'setor': 'Secos'},
    {'produto': 'Margarina',    'validade': date(2025, 5, 18), 'setor': 'Laticínios'},
]

# filter + lambda: manter só os que vencem em 7 dias ou menos
criticos = list(filter(
    lambda p: (p['validade'] - hoje).days <= 7,
    produtos_loja
))

print(f"📅 Hoje: {hoje.strftime('%d/%m/%Y')}")
print(f"🚨 PRODUTOS CRÍTICOS (vencem em até 7 dias)")
print("-" * 50)
for p in criticos:
    dias = (p['validade'] - hoje).days
    alerta = "⚠️ AMANHÃ!" if dias <= 1 else f"{dias} dias"
    print(f"   {p['produto']:15} → {p['validade'].strftime('%d/%m/%Y')}  ({alerta})")

print(f"\n   Total: {len(criticos)} produtos para retirar das prateleiras")
````

````python
# ── DEMONSTRAÇÃO 3B ── Múltiplos filtros encadeados ───────────────────

# Só laticínios críticos
laticinios_criticos = list(filter(
    lambda p: (p['validade'] - hoje).days <= 7 and p['setor'] == 'Laticínios',
    produtos_loja
))

# Produtos que NÃO estão vencendo (estoque seguro)
seguros = list(filter(
    lambda p: (p['validade'] - hoje).days > 30,
    produtos_loja
))

print("🥛 LATICÍNIOS CRÍTICOS")
for p in laticinios_criticos:
    print(f"   {p['produto']}  →  {p['validade']}")

print()
print("✅ ESTOQUE SEGURO (mais de 30 dias)")
for p in seguros:
    print(f"   {p['produto']}  →  {p['validade']}")
````

## 📦 SEÇÃO 4 — Reajuste de Preços
### Conceito: Lambda com `map()`

**Situação:** O fornecedor avisou que haverá um reajuste de 8% em todos os  
produtos do setor de Padaria. O sistema precisa atualizar todos os preços  
de uma vez sem um loop manual produto por produto.

---

### 📖 Teoria

`map(função, lista)` aplica a função a **cada elemento** da lista e retorna a lista transformada.

```python
map(lambda x: transformação, lista)
#           ↑ recebe cada item
#                       ↑ retorna o item transformado
```

> **Analogia:** map é como uma esteira de fábrica. Lambda é a máquina que transforma cada item.

````python
# ── DEMONSTRAÇÃO 4A ── Reajuste de preços ────────────────────────────

padaria = [
    {'produto': 'Pão francês',   'preco': 0.75},
    {'produto': 'Bolo de fubá',  'preco': 18.90},
    {'produto': 'Croissant',     'preco': 6.50},
    {'produto': 'Pão de queijo', 'preco': 2.30},
    {'produto': 'Brioche',       'preco': 14.00},
]

REAJUSTE = 0.08  # 8%

# map + lambda: criar nova lista com preços reajustados
padaria_reajustada = list(map(
    lambda p: {**p, 'preco': round(p['preco'] * (1 + REAJUSTE), 2)},
    padaria
))

print("🥖 REAJUSTE DE PREÇOS — PADARIA (+8%)")
print(f"  {'Produto':20} {'Antes':>10} {'Depois':>10} {'Diferença':>10}")
print("  " + "-" * 52)
for antes, depois in zip(padaria, padaria_reajustada):
    diff = depois['preco'] - antes['preco']
    print(f"  {antes['produto']:20} R$ {antes['preco']:6.2f}   R$ {depois['preco']:6.2f}   +R$ {diff:.2f}")
````

````python
# ── DEMONSTRAÇÃO 4B ── map com lista simples e múltiplas transformações

precos = [10.00, 25.50, 8.75, 42.00, 3.99]

# Aplicar desconto de 10%
com_desconto = list(map(lambda x: round(x * 0.90, 2), precos))

# Converter para centavos (sistema legado)
em_centavos = list(map(lambda x: int(x * 100), precos))

# Formatar como string monetária
formatados = list(map(lambda x: f"R$ {x:.2f}", precos))

print("💰 TRANSFORMAÇÕES COM map + lambda")
print(f"  Original:       {precos}")
print(f"  Com 10% off:    {com_desconto}")
print(f"  Em centavos:    {em_centavos}")
print(f"  Formatados:     {formatados}")
````

## 📦 SEÇÃO 5 — Total do Carrinho
### Conceito: Lambda com `reduce()`

**Situação:** Um cliente chegou ao caixa com vários itens no carrinho.  
O sistema precisa somar todos os preços para chegar ao total da compra.  
`reduce()` faz isso acumulando um resultado sobre toda a lista.

---

### 📖 Teoria

`reduce(função, lista)` aplica a função **acumulativamente** — pega os dois primeiros,  
aplica a função, pega o resultado com o terceiro, e assim por diante:

```
lista = [10, 5, 3, 8]

reduce(lambda acc, x: acc + x, lista)
  → (10 + 5) = 15
  → (15 + 3) = 18
  → (18 + 8) = 26  ✅
```

> **Analogia:** reduce é como um caixa somando item por item num papel — guarda o subtotal e adiciona o próximo.

````python
# ── DEMONSTRAÇÃO 5A ── Somando o carrinho ────────────────────────────

carrinho = [
    {'produto': 'Arroz 5kg',       'preco': 28.90, 'qtd': 2},
    {'produto': 'Feijão 1kg',      'preco': 9.50,  'qtd': 3},
    {'produto': 'Óleo de soja',    'preco': 7.80,  'qtd': 1},
    {'produto': 'Macarrão',        'preco': 4.20,  'qtd': 4},
    {'produto': 'Molho de tomate', 'preco': 3.50,  'qtd': 2},
]

# Primeiro: calcular subtotal de cada item com map
subtotais = list(map(lambda item: item['preco'] * item['qtd'], carrinho))

# Depois: somar tudo com reduce
total = reduce(lambda acc, x: acc + x, subtotais)

print("🛒 CARRINHO DE COMPRAS")
print("-" * 50)
for item, sub in zip(carrinho, subtotais):
    print(f"  {item['produto']:22} {item['qtd']}x R$ {item['preco']:.2f} = R$ {sub:.2f}")
print("-" * 50)
print(f"  {'TOTAL':>43} R$ {total:.2f}")
````

````python
# ── DEMONSTRAÇÃO 5B ── Reduce para encontrar o maior preço ───────────

precos_lista = [28.90, 9.50, 7.80, 4.20, 3.50]

# Encontrar o maior preço sem usar max()
maior = reduce(lambda a, b: a if a > b else b, precos_lista)

# Encontrar o menor
menor = reduce(lambda a, b: a if a < b else b, precos_lista)

print("📊 ANÁLISE DE PREÇOS COM reduce")
print(f"   Preços:       {precos_lista}")
print(f"   Maior preço:  R$ {maior:.2f}")
print(f"   Menor preço:  R$ {menor:.2f}")
print()
print("ℹ️  Na prática, use max() e min() — reduce é para lógica customizada!")
````

## 📦 SEÇÃO 6 — Programa de Fidelidade
### Conceito: Lambda com `apply()` no pandas

**Situação:** O mercado lançou um programa de fidelidade com 3 categorias de cliente:  
**Bronze** (até R$500/mês), **Prata** (R$501–R$1500) e **Ouro** (acima de R$1500).  
O setor de marketing precisa categorizar todos os clientes no DataFrame.

---

### 📖 Teoria

`df['coluna'].apply(lambda x: ...)` aplica uma função a cada valor de uma coluna.  
Quando precisar usar **duas colunas**, passe `axis=1` para receber a linha inteira:

```python
df.apply(lambda row: row['col1'] + row['col2'], axis=1)
```

````python
# ── DEMONSTRAÇÃO 6A ── Categorizar clientes ───────────────────────────

clientes = pd.DataFrame({
    'nome':          ['Ana Lima', 'Bruno Costa', 'Carla Silva', 'Diego Melo', 'Eva Santos'],
    'gasto_mensal':  [320.00, 1850.00, 780.00, 450.00, 2200.00],
    'meses_cliente': [6, 24, 12, 3, 36],
})

# Lambda com if/else encadeado para categorizar
clientes['categoria'] = clientes['gasto_mensal'].apply(
    lambda g: 'Ouro' if g > 1500 else ('Prata' if g > 500 else 'Bronze')
)

# Lambda para calcular gasto total histórico
clientes['gasto_total'] = clientes.apply(
    lambda row: row['gasto_mensal'] * row['meses_cliente'],
    axis=1
)

# Lambda para gerar mensagem personalizada
clientes['mensagem'] = clientes.apply(
    lambda row: f"Olá {row['nome'].split()[0]}! Você é cliente {row['categoria']}.",
    axis=1
)

print("🏆 PROGRAMA DE FIDELIDADE — CLIENTES CATEGORIZADOS")
print(clientes[['nome', 'gasto_mensal', 'categoria', 'gasto_total']].to_string(index=False))
print()
print("📱 MENSAGENS PERSONALIZADAS")
for msg in clientes['mensagem']:
    print(f"   {msg}")
````

### Conceito: Lambda em limpeza e normalização

**Situação:** Os nomes dos produtos foram cadastrados por vários funcionários  
de formas inconsistentes: maiúsculas, minúsculas, espaços sobrando, abreviações erradas.  
O sistema de BI não consegue agrupar corretamente. É hora de limpar os dados.

---

### 📖 Teoria

Lambda é ideal para limpeza quando a regra é simples e aplicada coluna a coluna.
Para regras complexas, prefira `def`. A dica de ouro:

```python
# Se cabe em uma linha legível → lambda
# Se precisa de mais de uma linha → def
```

````python
# ── DEMONSTRAÇÃO 7 ── Limpeza de dados de produtos ────────────────────
import re

produtos_sujos = pd.DataFrame({
    'nome_bruto': [
        '  ARROZ TIPO 1  ',
        'feijão carioca',
        'OLEO   DE SOJA',
        '  macarrao espaguete  ',
        'MOLHO    tomate'
    ],
    'preco_str': ['R$28,90', 'R$ 9.50', 'R$7,80', ' R$ 4,20 ', 'R$3.50'],
    'codigo':    ['  ARR-001 ', 'FEJ 002', '  OLE003', 'MAC-004  ', 'MOL 005'],
})

# 1. Normalizar nome: strip + title case + espaços duplos
produtos_sujos['nome'] = produtos_sujos['nome_bruto'].apply(
    lambda x: re.sub(r' +', ' ', x.strip().title())
)

# 2. Converter preço string para float
produtos_sujos['preco'] = produtos_sujos['preco_str'].apply(
    lambda x: float(x.strip().replace('R$', '').replace(' ', '').replace(',', '.'))
)

# 3. Padronizar código: strip + uppercase + remover espaços internos
produtos_sujos['codigo_limpo'] = produtos_sujos['codigo'].apply(
    lambda x: re.sub(r'\s', '', x.strip()).upper()
)

print("🧹 DADOS ANTES E DEPOIS DA LIMPEZA")
print()
colunas = ['nome_bruto', 'nome', 'preco_str', 'preco', 'codigo', 'codigo_limpo']
for _, row in produtos_sujos.iterrows():
    print(f"  ANTES:  nome='{row['nome_bruto']}'  |  preco='{row['preco_str']}'  |  cod='{row['codigo']}'")
    print(f"  DEPOIS: nome='{row['nome']}'  |  preco=R${row['preco']:.2f}  |  cod='{row['codigo_limpo']}'")
    print()
````

### Conceito: Lambda com `groupby()` no pandas

**Situação:** O diretor quer um relatório de performance de cada departamento:

Lambda dentro de `groupby().agg()` permite calcular métricas customizadas por grupo.

### 📖 Teoria

```python

#                                          ↑ g é a Series do grupo
```

````python
# ── DEMONSTRAÇÃO 8 ── Relatório por departamento ──────────────────────

vendas = pd.DataFrame({
    'departamento': ['Padaria','Padaria','Frios','Frios','Frios','Secos','Secos','Laticínios','Laticínios','Laticínios'],
    'produto':      ['Pão','Bolo','Presunto','Queijo','Salame','Arroz','Feijão','Leite','Iogurte','Manteiga'],
    'preco':        [0.75, 18.90, 32.00, 28.50, 45.00, 28.90, 9.50, 6.80, 4.50, 9.80],
    'qtd_vendida':  [320, 45, 18, 30, 12, 95, 80, 210, 140, 75],
})

vendas['faturamento'] = vendas['preco'] * vendas['qtd_vendida']

# Relatório com métricas customizadas via lambda
relatorio = vendas.groupby('departamento').agg(
    total_faturamento=('faturamento', lambda g: round(g.sum(), 2)),
    ticket_medio=     ('preco',       lambda g: round(g.mean(), 2)),
    produto_destaque= ('produto',     lambda g: vendas.loc[g.index, 'faturamento'].idxmax()),
    qtd_total=        ('qtd_vendida', lambda g: g.sum()),
).reset_index()

# Resolver o índice para o nome do produto
relatorio['produto_destaque'] = relatorio['produto_destaque'].apply(
    lambda idx: vendas.loc[idx, 'produto']
)

print("📊 RELATÓRIO DE PERFORMANCE POR DEPARTAMENTO")
print("-" * 70)
for _, row in relatorio.iterrows():
    print(f"  🏪 {row['departamento']}")
    print(f"     Faturamento:       R$ {row['total_faturamento']:,.2f}")
    print(f"     Ticket médio:      R$ {row['ticket_medio']:.2f}")
    print(f"     Produto destaque:  {row['produto_destaque']}")
    print(f"     Unidades vendidas: {row['qtd_total']}")
    print()
````

## 🏆 SEÇÃO 9 — Missão Final: Pipeline Completo

**Situação:** O diretor pediu um relatório completo de fim de mês.  
Você vai montar um **pipeline de dados** que usa todas as técnicas de lambda  
aprendidas nas seções anteriores — do raw data ao relatório final.

Execute as células em ordem e veja o pipeline completo funcionando.

````python
# ── PIPELINE COMPLETO ── Do dado bruto ao relatório ───────────────────

# 1️⃣  DADOS BRUTOS (simulando uma importação com problemas)
dados_brutos = pd.DataFrame({
    'produto':       ['  ARROZ 5KG  ', 'feijão 1kg', 'OLEO SOJA', '  macarrao  ', 'Molho Tomate'],
    'departamento':  ['Secos', 'secos', 'SECOS', 'Secos', 'secos'],
    'preco_str':     ['R$28,90', 'R$9,50', 'R$7,80', 'R$4,20', 'R$3,50'],
    'qtd_estoque':   [95, 80, 60, 140, 110],
    'qtd_vendida':   [45, 38, 20, 70, 55],
})

print("1️⃣  DADOS BRUTOS")
print(dados_brutos.to_string(index=False))
print()
````

````python
# 2️⃣  LIMPEZA com lambda + apply
df = dados_brutos.copy()

df['produto']      = df['produto'].apply(lambda x: x.strip().title())
df['departamento'] = df['departamento'].apply(lambda x: x.strip().title())
df['preco']        = df['preco_str'].apply(
    lambda x: float(x.replace('R$','').replace(',','.'))
)

print("2️⃣  APÓS LIMPEZA")
print(df[['produto','departamento','preco','qtd_estoque','qtd_vendida']].to_string(index=False))
print()
````

````python
# 3️⃣  ENRIQUECIMENTO com map e apply

df['faturamento']     = df.apply(lambda row: round(row['preco'] * row['qtd_vendida'], 2), axis=1)
df['estoque_restante']= df.apply(lambda row: row['qtd_estoque'] - row['qtd_vendida'], axis=1)
df['giro']            = df.apply(
    lambda row: 'Alto' if row['qtd_vendida'] / row['qtd_estoque'] > 0.6 else
                ('Médio' if row['qtd_vendida'] / row['qtd_estoque'] > 0.3 else 'Baixo'),
    axis=1
)

print("3️⃣  APÓS ENRIQUECIMENTO")
print(df[['produto','preco','faturamento','estoque_restante','giro']].to_string(index=False))
print()
````

````python
# 4️⃣  ANÁLISE FINAL com filter, sorted e reduce

registros = df.to_dict('records')

# Produtos com giro alto (filter)
giro_alto = list(filter(lambda p: p['giro'] == 'Alto', registros))

# Ordenar por faturamento (sorted)
por_fat = sorted(registros, key=lambda p: p['faturamento'], reverse=True)

# Total geral (reduce)
total_geral = reduce(lambda acc, p: acc + p['faturamento'], registros, 0)

print("4️⃣  RELATÓRIO FINAL")
print("=" * 50)
print(f"  Total de produtos: {len(registros)}")
print(f"  Faturamento total: R$ {total_geral:,.2f}")
print()
print("  🥇 RANKING POR FATURAMENTO")
for i, p in enumerate(por_fat, 1):
    print(f"     {i}. {p['produto']:20} R$ {p['faturamento']:6.2f}  [{p['giro']}]")
print()
print("  🔥 PRODUTOS COM GIRO ALTO")
for p in giro_alto:
    pct = round(p['qtd_vendida'] / p['qtd_estoque'] * 100)
    print(f"     {p['produto']:20} {pct}% vendido")
print("=" * 50)
print("  ✅ Relatório gerado com sucesso!")
````
