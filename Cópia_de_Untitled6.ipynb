# ----------------------------------------------------
# PARTE 1: NOSSO PRÓPRIO "SORTEADOR" DE NÚMEROS
# ----------------------------------------------------
# Como não podemos usar a biblioteca "random", vamos criar
# o nosso! Isso se chama LCG (Gerador Congruente Linear).
# É só uma conta matemática que produz números que PARECEM
# aleatórios. Não precisa entender 100% — só saber que funciona!

```python
semente = 12345 # Numero inicial ( Chama de "Seed" )

def sortear(maximo):
  #A palavra "Global" avisa que vamos mudar a viriavel de fora
  global semente
  #Formula matematica que embaralha o numero
  semente = (semente * 1103515245 + 12345) % 2147483648
  #0 operador % ( resto da divisão) gerante que o resultado fique entre 0 e ( maximo -1)
  return semente % maximo
```

# ----------------------------------------------------
# PARTE 2: CONFIGURAÇÕES INICIAIS
# ----------------------------------------------------

```python
Largura = 10 #quantas colunas o tabuleiro terá
Altura = 8 #quantas linhas o tabuleiro terá

#A cobriha é uma LISTA de posiçoes [coluna, linha]
#Começa com 1 pedacinho no meio do tabuleiro
cobra = [[5, 4]]

# Direção inicial da cobra ( começa até o jogador agir)
direcao = "parado"

#posição inicial da comida
comida_x = sortear(Largura)
comida_y = sortear(Altura)

#Variavel que controla se o jogo continua rodando
#Aqui já aparece logica boleana: True = False
jogo_rodando = True

#Pontuação do jogador
pontos = 0
```

# ----------------------------------------------------
# PARTE 3: FUNÇÃO QUE DESENHA O TABULEIRO
# ----------------------------------------------------

```python
def desenhar_tabuleiro():
  #imprime varias linhas em branco para "limpar" a tela antiga
  print("\n" * 3)

  print("=" * (Largura * 2 + 3)) #Borda de cima

  # Percorremos cada linha (y) e cada coluna (x) do tabuleiro
  for y in range(Altura):
    linha = " | "
    for x in range(Largura):

      if [x ,y] == cobra[0]:
        #e a cabeça da cobra
        linha += "@"
      elif [x, y] in cobra:
        # é um pedaço do corpo
        linha += "o"
      elif x == comida_x and y == comida_y:
        # é a Comida
        linha += "* "
      else:
        # senão é espaço vazio
        linha +=". "

    linha += " | " #Borda da direita
    print(linha)

  print("=" * (Largura * 2 + 3)) #Borda de baixo
  print("Pontos: " + str(pontos))
```

# ----------------------------------------------------
# PARTE 4: FUNÇÃO QUE MOVE A COBRA
# ----------------------------------------------------

```python
def mover_cobra():
  # Pegamos a posição da cabeça
  cabeca_x = cobra[0][0]
  cabeca_y = cobra[0][1]

  # IF/ELIF/ELSE para decidir para onde a cabeça vai
  if direcao == "Cima":
    cabeca_y = cabeca_y - 1
  elif direcao == "Baixo":
    cabeca_y = cabeca_y + 1
  elif direcao == "Esquerda":
    cabeca_x = cabeca_x - 1
  elif direcao == "Direita":
    cabeca_x = cabeca_x + 1
  else:
    #se direção for "parado", a função termina aqui
    return

  #Cria a nova cabeça e coloca o inicio da lista
  nova_cabeca = [cabeca_x, cabeca_y]
  cobra.insert(0, nova_cabeca)

```

# ----------------------------------------------------
# PARTE 5: FUNÇÃO QUE VERIFICA SE COMEU A COMIDA
# ----------------------------------------------------

```python
def verificar_comida():
  # Avisamos ao python que vamos MUDAR essas variaveis globais
  global comida_x, comida_y, pontos

  cabeca = cobra [0]

  #IF: a cabeça esta no mesmo lugar da comida
  if cabeca[0] == comida_x and cabeca[1] == comida_y:
    #Se Sim (True): aumenta pontos e sorteia nova comida
    pontos = pontos + 1
    print (">>> Voce comeu! +1 ponto <<<")

    #Sorteio nova posição usando sorteador
    comida_x = sortear(Largura)
    comida_y = sortear(Altura)
  else:
    # se Não comeu (false): remove o ulimo pedacinho
    # Assim a cobra "anda" sem crescer
    cobra.pop()
```

# ----------------------------------------------------
# PARTE 6: FUNÇÃO QUE VERIFICA SE O JOGADOR PERDEU
# ----------------------------------------------------

```python
def verificar_fim_de_jogo():
  global jogo_rodando

  cabeca_x = cobra[0][0]
  cabeca_y = cobra[0][1]

  #Condição 1: bateu na parede?
  #o "or" significa OU: basta uma das condições ser True
  if cabeca_x < 0 or cabeca_x >= Largura or cabeca_y < 0 or cabeca_y >= Altura:
    print("")
    print("*** GAME OVER: Voce bateu na parede! ***")
    jogo_rodando = False # Mudamos True -> Falsepara parar o jogo
    return

  #Condição 2: Bateu no proprio corpo?
  # cobra[1:] significa "todos os pedaços, exceto cabeça"

  if [cabeca_x, cabeca_y] in cobra[1:]:
    print("")
    print("*** game over: Voce se mordeu! ***")
    jogo_rodando = False
```

# ----------------------------------------------------
# PARTE 7: LOOP PRINCIPAL DO JOGO
# ----------------------------------------------------

```python
print("=" * 30)
print(" Jogo da Cobrinha")
print("=" * 30)
print("Comandos:")
print(" W = cima")
print(" S = baixo")
print(" A = esquerda")
print(" d = direita")
print(" q = para sair")
print("=" * 30)

#While roda enquanto a condição for True
#QUando jogo_Rodando virar Falso, o loop para sozinho
while jogo_rodando:

  desenhar_tabuleiro()

  #pede um comando para o jogador
  # . Lower() transforma em minusculo ( W vira w)
  comando = input("Move (w/a/s/d) ou q para sair").lower()

  #IF/ Elif/ Else para interromper o comando do jogador
  if comando == "w":
    direcao = "Cima"
  elif comando == "s":
    direcao = "Baixo"
  elif comando == "a":
    direcao = "Esquerda"
  elif comando == "d":
    direcao = "Direita"
  elif comando == "q":
    print("Voce desistiu! Até a proxima!")
    jogo_rodando = False
    break #Sair do loop imediatamente

  else:
    print("Comando Invalido! Use w,a,s,d ou q." )
    continue # Volta para o inicio do loop sem mover

  mover_cobra()
  verificar_fim_de_jogo()

  if jogo_rodando:
    verificar_comida()

print("")
print("Fim de jogo Pontuacao final: " + str(pontos))
```
