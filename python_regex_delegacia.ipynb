### 📚 Referência Rápida — Cola do Analista

| Símbolo | Significado | Exemplo |
|---------|-------------|--------|
| `\d` | Qualquer dígito (0-9) | `\d{4}` → `2024` |
| `\w` | Letra, dígito ou _ | `\w+` → `user_01` |
| `\s` | Espaço, tab, newline | `\s+` → `   ` |
| `.` | Qualquer caractere | `c.t` → `cat`, `cot` |
| `+` | 1 ou mais vezes | `\d+` → `42`, `999` |
| `*` | 0 ou mais vezes | `\d*` → `` ou `99` |
| `?` | Opcional (0 ou 1) | `-?` → `` ou `-` |
| `{n}` | Exatamente n vezes | `\d{3}` → `123` |
| `{n,m}` | De n a m vezes | `\d{2,4}` → `12`, `1234` |
| `[abc]` | a, b ou c | `[aeiou]` → vogais |
| `[A-Z]` | Letra maiúscula | `[A-Z][a-z]+` → `Maria` |
| `[^x]` | Qualquer coisa exceto x | `[^0-9]` → não-dígito |
| `\.` | Ponto literal | `\d{3}\.\d{3}` → `123.456` |
| `^` | Início da string | `^\d` → começa com número |
| `$` | Fim da string | `\d$` → termina com número |
| `(grupo)` | Captura grupo | `(\d{2})/(\d{4})` |
| `a\|b` | a ou b | `jpg\|png\|gif` |

```python
# ✅ Execute esta célula primeiro — ela carrega o sistema de testes
import re

missoes_resolvidas = []

def testar(missao, obtido, esperado, mensagem_ok=""):
    print(f"\n{'='*55}")
    print(f"  MISSÃO {missao}")
    print(f"{'='*55}")
    if obtido == esperado:
        missoes_resolvidas.append(missao)
        print(f"  ✅ CASO ENCERRADO!")
        if mensagem_ok:
            print(f"  📋 {mensagem_ok}")
        print(f"  Resultado: {obtido}")
    else:
        print(f"  ❌ INVESTIGAÇÃO AINDA EM ANDAMENTO...")
        print(f"  Seu resultado : {obtido}")
        print(f"  Esperado      : {esperado}")
    print(f"{'='*55}")

def placar():
    total = 10
    resolvidas = len(set(missoes_resolvidas))
    barra = '█' * resolvidas + '░' * (total - resolvidas)
    print(f"\n🏆 PLACAR DA DELEGACIA")
    print(f"   [{barra}] {resolvidas}/{total} missões")
    if resolvidas == total:
        print("   🎉 OPERAÇÃO CÓDIGO LIMPO ENCERRADA COM SUCESSO!")
    elif resolvidas >= 7:
        print("   🔥 Excelente trabalho, analista!")
    elif resolvidas >= 4:
        print("   💪 Bom progresso! Continue investigando.")
    else:
        print("   🔎 Continue, o caso depende de você!")

print("✅ Sistema de testes carregado. Boa investigação!")
```

## 🗂️ MISSÃO 01 — O Informante

**Situação:** Um informante infiltrado na organização criminosa enviou uma mensagem codificada ao delegado.  
O texto parece um e-mail comum, mas escondido entre as palavras está um número de contato secreto.  
Cada dígito foi separado por traços para dificultar a leitura.

**Sua tarefa:** Extraia todos os dígitos da mensagem para montar o número de contato.

**Conceito treinado:** `\d` — captura qualquer dígito de 0 a 9

```python
mensagem = "Reuniao as 1-4h no ap. 3, bloco 7. Leve os documentos. Cod: 9-8-6-5"

# 👇 Substitua SEU_PADRAO pelo padrão correto
padrao = r"SEU_PADRAO"

resultado = re.findall(padrao, mensagem)

testar(
    "01",
    resultado,
    ['1', '4', '3', '7', '9', '8', '6', '5'],
    "Dígitos extraídos! O contato pode ser rastreado."
)
```

```python
# 💡 DICA — Execute se precisar de ajuda
print("""
💡 DICA MISSÃO 01
─────────────────────────────────────────
  \\d  →  captura qualquer dígito (0-9)

  Exemplo:
    re.findall(r'\\d', 'abc 3 def 7')  →  ['3', '7']

  Sem quantificador, \\d captura 1 dígito por vez.
  Para capturar números inteiros, use \\d+
─────────────────────────────────────────
""")
```

```python
# 🔓 GABARITO — Só olhe depois de tentar!
print("""
🔓 GABARITO MISSÃO 01
─────────────────────────────────────────
  padrao = r'\\d'

  \\d captura qualquer dígito individualmente.
  re.findall retorna uma lista com cada dígito.
─────────────────────────────────────────
""")
```

## 🗂️ MISSÃO 02 — E-mails dos Cúmplices

**Situação:** Você interceptou um dump do servidor da organização.  
O arquivo está corrompido — mistura de logs de erro, lixo e, no meio,  
os endereços de e-mail de dois cúmplices que trocaram mensagens suspeitas.  
A promotoria precisa desses e-mails para emitir o mandado.

**Sua tarefa:** Extraia apenas os e-mails válidos do log corrompido.

**Conceito treinado:** Classes de caracteres `[\w.]+`, arroba `@`, e extensões `\.[a-z]{2,}`

```python
log = "ERR_404 timestamp:2024 | joao.silva@fbi.gov falha_auth | xpto@@invalido.com | maria_99@policia.sp.gov.br | --END--"

# 👇 Substitua SEU_PADRAO pelo padrão correto
padrao = r"SEU_PADRAO"

resultado = re.findall(padrao, log)

testar(
    "02",
    resultado,
    ['joao.silva@fbi.gov', 'maria_99@policia.sp.gov.br'],
    "Cúmplices identificados! Mandado emitido."
)
```

```python
# 💡 DICA — Execute se precisar de ajuda
print("""
💡 DICA MISSÃO 02
─────────────────────────────────────────
  Um e-mail tem 3 partes:
    1. nome do usuário  →  [\\w.]+
    2. arroba           →  @
    3. domínio          →  [\\w.]+\\.[a-z]{2,}

  Monte as 3 partes juntas no padrão.
  Teste: joao.silva@fbi.gov deve casar.
         xpto@@invalido.com NÃO deve casar (dois @@).
─────────────────────────────────────────
""")
```

```python
# 🔓 GABARITO
print("""
🔓 GABARITO MISSÃO 02
─────────────────────────────────────────
  padrao = r'[\\w.]+@[\\w.]+\\.[a-z]{2,}'

  [\\w.]+ → nome: letras, números, _ e ponto
  @       → arroba literal
  [\\w.]+ → domínio
  \\.[a-z]{2,} → extensão com pelo menos 2 letras
─────────────────────────────────────────
""")
```

## 🗂️ MISSÃO 03 — Câmeras de Segurança

**Situação:** As câmeras de segurança da cena do crime gravaram a passagem de vários veículos.  
O arquivo de log gerado pelo sistema mistura placas reais (formato `AAA-1234`) com dados  
corrompidos e ruídos. A perícia precisa das placas para cruzar com o banco de dados do DETRAN.

**Sua tarefa:** Extraia apenas as placas no formato antigo brasileiro: 3 letras maiúsculas + hífen + 4 dígitos.

**Conceito treinado:** `[A-Z]{3}` para letras, `-` literal, `\d{4}` para dígitos

```python
log_cameras = """
cam01 | 08:12 | ABC-1234 | velocidade_ok
cam02 | 08:15 | lixo-99  | erro_leitura
cam03 | 08:31 | XYZ-5678 | velocidade_ok
cam04 | 08:44 | 123-ABCD | formato_invalido
cam05 | 09:02 | QWE-9090 | velocidade_ok
"""

# 👇 Substitua SEU_PADRAO pelo padrão correto
padrao = r"SEU_PADRAO"

resultado = re.findall(padrao, log_cameras)

testar(
    "03",
    resultado,
    ['ABC-1234', 'XYZ-5678', 'QWE-9090'],
    "Placas encontradas! Cruzamento com DETRAN iniciado."
)
```

```python
# 💡 DICA
print("""
💡 DICA MISSÃO 03
─────────────────────────────────────────
  Placa antiga: 3 LETRAS MAIÚSCULAS + hífen + 4 DÍGITOS

  [A-Z]   → uma letra maiúscula
  {3}     → exatamente 3 vezes
  -       → hífen literal
  \\d{4}  → exatamente 4 dígitos

  Por que '123-ABCD' não deve casar?
  Porque começa com números, não letras!
─────────────────────────────────────────
""")
```

```python
# 🔓 GABARITO
print("""
🔓 GABARITO MISSÃO 03
─────────────────────────────────────────
  padrao = r'[A-Z]{3}-\\d{4}'
─────────────────────────────────────────
""")
```

## 🗂️ MISSÃO 04 — Linha do Tempo

**Situação:** O perito computacional recuperou um log do servidor da empresa investigada.  
Para reconstruir a **linha do tempo** do crime, o delegado precisa de todos os horários  
registrados no formato `HH:MM`. Entradas malformadas (como `9:5`) devem ser descartadas.

**Sua tarefa:** Extraia horários válidos com exatamente 2 dígitos em cada parte.

**Conceito treinado:** `\d{2}` para exigir exatamente 2 dígitos, `:` literal

```python
log_servidor = """
[INFO]  08:32  usuario=admin  login_ok
[WARN]  9:5    tentativa_falha  (formato inválido no sistema legado)
[ERR]   14:05  acesso_negado  arquivo=relatorio_secreto.pdf
[INFO]  23:59  usuario=admin  logout
[ERR]   7:00   conexao_recusada
[INFO]  01:15  backup_automatico_ok
"""

# 👇 Substitua SEU_PADRAO pelo padrão correto
padrao = r"SEU_PADRAO"

resultado = re.findall(padrao, log_servidor)

testar(
    "04",
    resultado,
    ['08:32', '14:05', '23:59', '01:15'],
    "Linha do tempo reconstruída! Relatório enviado ao juiz."
)
```

```python
# 💡 DICA
print("""
💡 DICA MISSÃO 04
─────────────────────────────────────────
  \\d{2}  → EXATAMENTE 2 dígitos (não mais, não menos)
  :       → dois pontos literal
  \\d{2}  → mais 2 dígitos para os minutos

  Por que '9:5' não deve casar?
  Porque \\d{2} exige 2 dígitos — '9' só tem 1.

  Por que '7:00' não deve casar?
  Mesma razão — '7' tem apenas 1 dígito.
─────────────────────────────────────────
""")
```

```python
# 🔓 GABARITO
print("""
🔓 GABARITO MISSÃO 04
─────────────────────────────────────────
  padrao = r'\\d{2}:\\d{2}'
─────────────────────────────────────────
""")
```

## 🗂️ MISSÃO 05 — O Depoimento Bagunçado

**Situação:** O suspeito foi levado à delegacia e seu depoimento foi digitado às pressas  
por um escrivão nervoso. O texto ficou cheio de espaços duplos e triplos.  
O sistema de arquivamento judicial **recusa documentos com espaços irregulares**.  
Você precisa limpar o depoimento antes de protocolá-lo.

**Sua tarefa:** Use `re.sub()` para substituir 2 ou mais espaços consecutivos por 1 espaço.

**Conceito treinado:** `re.sub()`, quantificador `{2,}` ou ` +`

```python
depoimento = "Eu  estava  em   casa  naquela   noite.  Não  saí  de  lá,  juro."

# 👇 Substitua SEU_PADRAO pelo padrão correto
padrao = r"SEU_PADRAO"

resultado = re.sub(padrao, " ", depoimento)

testar(
    "05",
    resultado,
    "Eu estava em casa naquela noite. Não saí de lá, juro.",
    "Depoimento normalizado e arquivado com sucesso!"
)
```

```python
# 💡 DICA
print("""
💡 DICA MISSÃO 05
─────────────────────────────────────────
  re.sub(padrao, substituto, texto)
  substitui cada match do padrão pelo substituto.

  Para 2 ou mais espaços:
    ' {2,}'  → espaço + {2,} significa 2 ou mais
    '  +'    → dois espaços + + significa 1 ou mais
              (ambos funcionam aqui)

  ⚠️  Não use \\s pois ele remove \\n também.
      Use espaço literal para ser preciso.
─────────────────────────────────────────
""")
```

```python
# 🔓 GABARITO
print("""
🔓 GABARITO MISSÃO 05
─────────────────────────────────────────
  padrao = r' {2,}'   # 2 ou mais espaços
  # ou
  padrao = r'  +'     # espaço + 1 ou mais espaços
─────────────────────────────────────────
""")
```

## 🗂️ MISSÃO 06 — Datas do Mandado

**Situação:** O promotor encontrou uma pilha de documentos digitalizados com datas misturadas.  
Alguns estão no formato correto `DD/MM/AAAA`, mas outros foram digitados sem zeros à esquerda  
(ex: `9/5/2024`). O sistema judicial só aceita datas com **exatamente 2 dígitos** no dia e mês.

**Sua tarefa:** Extraia apenas datas no formato estrito `DD/MM/AAAA`.

**Conceito treinado:** `\d{2}` para 2 dígitos, `/` literal, `\d{4}` para o ano

```python
documentos = """
  Contrato assinado em 25/12/2023.
  Reunião marcada: 01/07/2024.
  Rascunho criado em 9/5/2024 (sem zero — inválido).
  Prazo final: 31/10/2024.
  Versão antiga: 3/3/2020 (inválido).
"""

# 👇 Substitua SEU_PADRAO pelo padrão correto
padrao = r"SEU_PADRAO"

resultado = re.findall(padrao, documentos)

testar(
    "06",
    resultado,
    ['25/12/2023', '01/07/2024', '31/10/2024'],
    "Datas válidas extraídas para o processo!"
)
```

```python
# 💡 DICA
print("""
💡 DICA MISSÃO 06
─────────────────────────────────────────
  Formato: DD / MM / AAAA

  \\d{2}  → dia (exige 2 dígitos: '25', '01')
  /       → barra literal
  \\d{2}  → mês (exige 2 dígitos: '12', '07')
  /       → barra literal
  \\d{4}  → ano (exige 4 dígitos: '2024')

  '9/5/2024' não vai casar pois '9' e '5'
  têm apenas 1 dígito cada.
─────────────────────────────────────────
""")
```

```python
# 🔓 GABARITO
print("""
🔓 GABARITO MISSÃO 06
─────────────────────────────────────────
  padrao = r'\\d{2}/\\d{2}/\\d{4}'
─────────────────────────────────────────
""")
```

## 🗂️ MISSÃO 07 — CPFs dos Investigados

**Situação:** O cruzamento nacional de dados foi autorizado pelo juiz.  
Um arquivo com centenas de registros contém CPFs válidos e inválidos misturados.  
O sistema da Receita Federal exige CPFs **exatamente** no formato `NNN.NNN.NNN-NN`  
com pontos e hífen obrigatórios. Qualquer desvio deve ser descartado.

**Sua tarefa:** Extraia apenas CPFs no formato `NNN.NNN.NNN-NN`.

**Conceito treinado:** `\.` para ponto literal, `-` literal, `\d{3}` e `\d{2}`

```python
arquivo_receita = """
  REG_001 | Nome: João Silva    | CPF: 123.456.789-09 | Status: ativo
  REG_002 | Nome: Maria Santos  | CPF: 12.345.678-9   | Status: formato_invalido
  REG_003 | Nome: Carlos Lima   | CPF: 987.654.321-00 | Status: ativo
  REG_004 | Nome: Ana Souza     | CPF: 000.000.000-00 | Status: ativo
  REG_005 | Nome: Pedro Costa   | CPF: 111222333-44   | Status: sem_pontos
"""

# 👇 Substitua SEU_PADRAO pelo padrão correto
padrao = r"SEU_PADRAO"

resultado = re.findall(padrao, arquivo_receita)

testar(
    "07",
    resultado,
    ['123.456.789-09', '987.654.321-00', '000.000.000-00'],
    "CPFs válidos encontrados! Cruzamento na Receita Federal iniciado."
)
```

```python
# 💡 DICA
print("""
💡 DICA MISSÃO 07
─────────────────────────────────────────
  Estrutura do CPF:  NNN . NNN . NNN - NN

  \\d{3}  → 3 dígitos
  \\.     → ponto LITERAL (sem \ seria qualquer char!)
  \\d{3}  → mais 3 dígitos
  \\.     → ponto literal
  \\d{3}  → mais 3 dígitos
  -       → hífen literal
  \\d{2}  → 2 dígitos finais

  ⚠️  ATENÇÃO: . sem barra captura QUALQUER char!
      Use \\. para ponto literal.
─────────────────────────────────────────
""")
```

```python
# 🔓 GABARITO
print("""
🔓 GABARITO MISSÃO 07
─────────────────────────────────────────
  padrao = r'\\d{3}\\.\\d{3}\\.\\d{3}-\\d{2}'
─────────────────────────────────────────
""")
```

## 🗂️ MISSÃO 08 — Nomes Próprios nas Atas

**Situação:** As atas das reuniões secretas da organização foram recuperadas de um HD criptografado.  
Para montar o **organograma completo dos envolvidos**, o delegado precisa extrair todos os nomes próprios.  
Por convenção, nomes próprios começam com letra **maiúscula** seguida de letras **minúsculas**.

**Sua tarefa:** Extraia todas as palavras que começam com maiúscula seguidas de minúsculas.

**Conceito treinado:** `[A-Z]` para inicial maiúscula, `[a-z]+` para o restante

```python
ata = "Na reunião, Maria coordenou com Pedro e Carlos a entrega para Fernanda. O Mercado foi o ponto de encontro."

# 👇 Substitua SEU_PADRAO pelo padrão correto
padrao = r"SEU_PADRAO"

resultado = re.findall(padrao, ata)

testar(
    "08",
    resultado,
    ['Na', 'Maria', 'Pedro', 'Carlos', 'Fernanda', 'Mercado', 'Ok'],
    "Organograma montado! 6 envolvidos identificados."
)

# Nota: 'Na' e 'Ok' também casam por ser maiúscula + minúsculas
# Para filtrar só nomes, precisaria de lógica adicional
testar(
    "08",
    resultado,
    ['Na', 'Maria', 'Pedro', 'Carlos', 'Fernanda', 'Mercado', 'Ok'],
    "Organograma montado!"
)
```

```python
# 💡 DICA
print("""
💡 DICA MISSÃO 08
─────────────────────────────────────────
  Nomes próprios: inicial MAIÚSCULA + resto minúsculo

  [A-Z]    → uma letra maiúscula (A até Z)
  [a-z]+   → uma ou mais letras minúsculas

  Junte os dois: [A-Z][a-z]+

  Teste mental:
    'Maria'   → M (maiúsc) + aria (minúsc) ✅
    'pedro'   → p (minúsc) ❌ não casa
    'CARLOS'  → C (maiúsc) + ARLOS (maiúsc) ❌ não casa
─────────────────────────────────────────
""")
```

```python
# 🔓 GABARITO
print("""
🔓 GABARITO MISSÃO 08
─────────────────────────────────────────
  padrao = r'[A-Z][a-z]+'
─────────────────────────────────────────
""")
```

## 🗂️ MISSÃO 09 — Auditoria de Senhas

**Situação:** Após obter acesso ao servidor, você encontrou o banco de senhas dos funcionários.  
A política de segurança da empresa exige senhas entre **8 e 20 caracteres alfanuméricos**.  
Senhas fora desse padrão indicam contas comprometidas ou criadas irregularmente.

**Sua tarefa:** Use `re.fullmatch()` para classificar cada senha como válida ou inválida.

**Conceito treinado:** `re.fullmatch()`, `\w{8,20}` — o padrão deve cobrir a string INTEIRA

```python
senhas = {
    'admin':          'abc123',           # muito curta
    'joao.silva':     'senhaSegura1',      # válida
    'maria_99':       'x',                # muito curta
    'carlos.lima':    'MinhaSenha2024',    # válida
    'fernanda':       'senha_muito_longa_que_nao_deve_passar',  # muito longa
    'pedro.costa':    'Acesso_2024ok',     # válida
}

# 👇 Substitua SEU_PADRAO pelo padrão correto
padrao = r"SEU_PADRAO"

print("  AUDITORIA DE SENHAS")
print("  " + "-"*40)
classificacao = {}
for usuario, senha in senhas.items():
    valida = bool(re.fullmatch(padrao, senha))
    status = "✅ VÁLIDA  " if valida else "❌ INVÁLIDA"
    classificacao[usuario] = valida
    print(f"  {status} | {usuario:20} | {senha}")

print()
esperado = {
    'admin': False, 'joao.silva': True, 'maria_99': False,
    'carlos.lima': True, 'fernanda': False, 'pedro.costa': True
}
testar("09", classificacao, esperado, "Relatório de senhas gerado!")
```

```python
# 💡 DICA
print("""
💡 DICA MISSÃO 09
─────────────────────────────────────────
  re.fullmatch() exige que o padrão cubra
  a string INTEIRA (diferente de search/findall).

  \\w     → letra, dígito ou _
  {8,20}  → entre 8 e 20 caracteres

  Resultado:
    'abc123'       → 6 chars  ❌ (menos de 8)
    'senhaSegura1' → 12 chars ✅
    'senha_muito_longa...' → 36 chars ❌ (mais de 20)
─────────────────────────────────────────
""")
```

```python
# 🔓 GABARITO
print("""
🔓 GABARITO MISSÃO 09
─────────────────────────────────────────
  padrao = r'\\w{8,20}'
─────────────────────────────────────────
""")
```

## 🗂️ MISSÃO 10 — A Ficha Completa

**Situação:** O caso está quase encerrado. Você recebeu a ficha bruta do principal suspeito:  
um texto corrido com nome, CPF, telefone, e-mail e data do crime tudo junto.  
Para fechar o relatório final, você precisa extrair **cada dado separadamente**  
usando `re.search()` com os padrões que aprendeu nas missões anteriores.

**Sua tarefa:** Preencha os 4 padrões para extrair CPF, telefone, e-mail e data.

Agora que você domina o básico, explore:

- **Grupos nomeados:** `(?P<nome>\d+)` — acesse com `m.group('nome')`
- **Lookahead:** `\d+(?= reais)` — número seguido de 'reais' sem capturar
- **Lookbehind:** `(?<=R\$)\d+` — número precedido de 'R$'
- **Flags:** `re.IGNORECASE`, `re.MULTILINE`, `re.DOTALL`
- **re.compile():** compile o padrão uma vez para reuso eficiente
